# W3 - Activity-Diabetes

# Read the dataset to python environment

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
diabetes_data=pd.read_csv('diabetes_model.csv')
```

# Display the columns in the dataset

```python
diabetes_data.head()
```

# 1. Create a Count plot of Outcome 

```python
sns.countplot(x=diabetes_data['Outcome'])
```

# 2. Analyse the effect of age on Outcome using Boxplot

```python
sns.boxplot(x = 'Outcome', y = 'Age', data = diabetes_data)
plt.title('Effect of Age on Outcome')
```

# 3. Check for the chances of getting diabetes if it is common in family, using the Diabetes Pedigree Function 

```python
sns.violinplot(x = 'Outcome', y = 'DiabetesPedigreeFunction', data = diabetes_data)
```

# 4. Check for the relationship with Insulin & Glucose and group the datapoints by Outcome. 

```python
sns.relplot(x="Insulin", y="Glucose", data=diabetes_data, col="Outcome")
```

# 5. Create a heatmap on Correlation of features in the data.

```python
corr = diabetes_data.corr()
# print(corr)
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns)
```
